```python


# Step 1: Browse and upload the Excel file
uploaded = files.upload()  # Opens file picker

# Get the uploaded file name
file_name = list(uploaded.keys())[0]
print(f"Uploaded file: {file_name}")

# Step 2: Load the Excel file
df = pd.read_excel(file_name)

# Step 3: Check the first few rows
print(df.head())

# Step 4: Clean the data
df = df.dropna(subset=['Gender', 'Hb'])
df['Gender'] = df['Gender'].str.strip().str.capitalize()

# Step 5: Create Hb level bins
# Automatically determine bin edges using Freedman-Diaconis rule
q75, q25 = np.percentile(df['Hb'], [75 ,25])
iqr = q75 - q25
n = len(df['Hb'])
bin_width = 2 * iqr / n**(1/3)
num_bins = int(np.ceil((df['Hb'].max() - df['Hb'].min()) / bin_width))

# Create labels for bins
bins = np.linspace(df['Hb'].min(), df['Hb'].max(), num_bins + 1)
labels = [f"{round(bins[i],1)}-{round(bins[i+1],1)}" for i in range(len(bins)-1)]

# Bin the Hb level data
df['Hb_bin'] = pd.cut(df['Hb'], bins=bins, labels=labels, include_lowest=True)

# Step 6: Count number of males and females in each bin
gender_counts = df.groupby(['Hb_bin', 'Gender']).size().unstack(fill_value=0)
print(gender_counts)

# Step 7: Plot grouped bar chart
gender_counts.plot(kind='bar', figsize=(12,6), color=['skyblue','pink'])
plt.title('Hb Level Distribution by Gender')
plt.xlabel('Hb Level Range')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python


# Step 1: Browse and upload the Excel file
uploaded = files.upload()  # Opens file picker

# Get the uploaded file name
file_name = list(uploaded.keys())[0]
print(f"Uploaded file: {file_name}")

# Step 2: Load the Excel file
df = pd.read_excel(file_name)

# Step 3: Check the first few rows
print(df.head())

# Step 4: Clean the data - remove rows with missing Weight or Hb
df = df.dropna(subset=['Weight', 'Hb']).reset_index(drop=True)

# Step 5: Calculate correlation
correlation = df['Weight'].corr(df['Hb'])
print(f"Correlation between Weight and Hb: {correlation:.3f}")

# Step 6: Plot correlation heatmap
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Step 7: Scatter plot with regression line and equation
plt.figure(figsize=(8,6))
sns.scatterplot(x='Weight', y='Hb', data=df, color='blue')

# Fit a linear regression line
slope, intercept = np.polyfit(df['Weight'], df['Hb'], 1)
plt.plot(df['Weight'], slope*df['Weight'] + intercept, color='red')

# Add equation text on the plot
plt.text(min(df['Weight']), max(df['Hb']), f'y = {slope:.2f}x + {intercept:.2f}',
         color='red', fontsize=12, ha='left', va='top')

plt.title(f'Weight vs Hb Scatter Plot\nCorrelation r={correlation:.2f}')
plt.xlabel('Weight')
plt.ylabel('Hb Level')
plt.show()

plt.figure(figsize=(8,6))
plt.hexbin(df['Weight'], df['Hb'], gridsize=25, cmap='Blues', edgecolors='grey')
plt.colorbar(label='Count')
plt.xlabel('Weight')
plt.ylabel('Hb Level')
plt.title('Hexbin Plot: Weight vs Hb')
plt.show()

plt.figure(figsize=(8,6))
plt.hist2d(df['Weight'], df['Hb'], bins=20, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Weight')
plt.ylabel('Hb Level')
plt.title('2D Histogram: Weight vs Hb')
plt.show()

# Calculate frequency of each point
counts = df.groupby(['Weight', 'Hb']).size().reset_index(name='count')

plt.figure(figsize=(8,6))
plt.scatter(counts['Weight'], counts['Hb'], s=counts['count']*20, alpha=0.5, color='blue')
plt.xlabel('Weight')
plt.ylabel('Hb Level')
plt.title('Bubble Plot: Weight vs Hb')
plt.show()

```
